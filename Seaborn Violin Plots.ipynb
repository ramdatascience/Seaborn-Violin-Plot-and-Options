## Violin Plot Examples - Titanic Dataset

```python
#The violin plots provide much more information about the data as compared to the box plot.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv("tit_train.csv")
dataset.head()
```

```python
# PLOT : 1
#  Instead of plotting the quartile , the violin plot allows us to see all the components that actually corresponds to the data.
#  The area where the violin plot is thicker has a higher number of instances of age.
# In the plot below for males , its clearnly evident that passengers with age between 20 and 40 is higher than all the rest of the age brackets.
```

```python
sns.violinplot(x = 'Sex',y= 'Age',data=dataset )
```

```python
# Plot 2
# By using the HUE parameter, we add another categorical variable which provides lot more information to the plot.
```

```python
sns.violinplot(x='Sex', y ='Age',data=dataset,hue='Survived')
```

```python
# Plot 3
# Instead of plotting two difference graphs for the passengers who survived and those who didn't we can have 
# one violin plot, divided in to two halves, where one half represent surviving statistics and the other half represent the non-surviving passengers. 
#To do so we used the SPLIT parameter.
```

```python
sns.violinplot(x='Sex',y='Age',data=dataset,hue='Survived',split=True)
```
